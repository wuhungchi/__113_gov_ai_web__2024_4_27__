```python
from pydantic import BaseModel , Field,field_validator#field_validator為手動條件驗證
class Records(BaseModel):
    公司名稱:str = Field(alias='Name_Ins')
    登記編號:str = Field(alias='No')
    地址:str = Field(alias='Address')
    統編:int = Field(alias='tax_ID_number')
    緯度:float = Field(alias='wgs84aX')
    經度:float = Field(alias='wgs84aY')
    @field_validator("緯度","經度",mode='before')
    @classmethod
    def abc(cls, value):
        if value == '':
            return 0.0
        else:
            return value
class Result(BaseModel):
    records:list[Records]#要抓取的資料都在records裡面,所以必須將已定義為Records的records再定義為list

import json
from pprint import pprint
from pydantic import BaseModel,Field,field_validator
from pydantic import BaseModel , Field,field_validator

with open('新北市食品工廠清冊.json',encoding='utf-8')as file:#用open叫出檔案並定義為file
    json_content:str = file.read()#用read()來讀file並將它定義為json檔
    all_data = json.loads(json_content)#再用json.loads把json檔轉換成python檔
    
result:Result = Result.model_validate(all_data['result'])#result為最終想讀取的本體,用model_validate

for rec in result.records:
    print(rec)
```
