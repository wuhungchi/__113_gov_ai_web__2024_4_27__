```python
from csv import DictReader #用from import叫出DictReader語法來抓取資料
with open("aqx_p_488-2.csv",encoding="utf-8") as file:
    reader:DictReader = DictReader(file)
    sites: list[dict]= list(reader)#將資料轉換成list才能無限讀取,以免原本的DictReader只能讀一遍
    

for site in sites:
    
    print(site['sitename'],end=" ")
    print(site['county'],end=" ")
    print(site['aqi'],end=" ")
    print(site['status'])
```
